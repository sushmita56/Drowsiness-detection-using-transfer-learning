```python
trd = []
def get_training_data():    
    dataDirectory = 'data/train'
    img_size = 100
    for category in categories: 
        folder = os.path.join(dataDirectory, category)
        class_label = categories.index(category)
        for img in os.listdir(folder):  
            img_path = os.path.join(folder, img)
            img_arr = cv2.imread(img_path)
            image = cv2.resize(img_arr, (img_size, img_size))
            trd.append([img_arr, class_label])
```

```python
get_training_data()
```

```python
len(trd)
```

```python
import random
random.shuffle(trd)
```

```python
X = []
y = []
for features, labels in trd:
  X.append(features)
  y.append(labels)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model = Sequential()

model.add(InputLayer(input_shape=(128,128,3)))

model.add(Conv2D(64, (3,3), strides=(1,1), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(Conv2D(32, (3,3), strides=(1,1), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(Flatten())
model.add(Dense(50, activation='relu'))
model.add(Dense(4, activation='softmax'))
```

```python
model.compile(loss= 'binary_cross_entropy',
              optimizer = 'adam',
              metrics= ['accuracy'])
```

```python
model.fit(X, y, epochs=2, validation_split=0.1, batch_size=32)
```
